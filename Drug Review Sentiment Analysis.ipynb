```python
import nltk
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
df_lvg['review'] =  df_lvg.review.astype(str).str.replace('\[|\]|\'', '')
```

```python
df_lvg['review'] = df_lvg['review'].apply(lambda x: tokenizer.tokenize(x.lower()))
```

```python
df_lvg['review'].head()
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words
```

```python
df_lvg['review'] = df_lvg['review'].apply(lambda x: remove_stopwords(x))
```

```python
df_lvg['review'].head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
df_lvg['review'] =  df_lvg.review.astype(str).str.replace('\[|\]|\'', '')
df_lvg['review'] = df_lvg['review'].apply(lambda x: wordnet_lemmatizer.lemmatize(x))
```

```python
df_lvg.head()
```

```python
df_lvg.isnull().sum()
```

```python
for i in df_lvg['rating']:
    float(i)
```

```python
df_lvg["is_bad_review"] = df_lvg["rating"].apply(lambda x: 1 if x < 5 else 0)
```

```python
df_lvg.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
#df_lvg['text_as_string'] = df_lvg['review'].str[0].str.join(" ")
df_lvg['text_as_string'] =  df_lvg.review.astype(str).str.replace('\[|\]|\'', '')
df_lvg["sentiments"] = df_lvg['text_as_string'].apply(lambda x: sid.polarity_scores(x))
```

```python
df_lvg.head()
```

```python
df_lvg = pd.concat([df_lvg.drop(['sentiments'], axis=1), df_lvg['sentiments'].apply(pd.Series)], axis=1)
```

```python
df_lvg.head()
```

```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
# create doc2vec vector columns
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(df_lvg["text_as_string"].apply(lambda x: x.split(" ")))]
```

```python
# train a Doc2Vec model with our text data
model = Doc2Vec(documents, vector_size=5, window=2, min_count=1, workers=4)
```

```python
# transform each document into a vector data
doc2vec_df = df_lvg["text_as_string"].apply(lambda x: model.infer_vector(x.split(" "))).apply(pd.Series)
doc2vec_df.columns = ["doc2vec_vector_" + str(x) for x in doc2vec_df.columns]
reviews_df = pd.concat([df_lvg, doc2vec_df], axis=1)
```

```python
reviews_df.head()
```

```python
reviews_df["is_bad_review"].value_counts(normalize = True)
```

```python
# highest positive sentiment reviews
reviews_df.sort_values("pos", ascending = False)[["review", "pos"]].head(10)
```

```python
# lowest negative sentiment reviews
reviews_df.sort_values("neg", ascending = False)[["review", "neg"]].head(10)
```

```python
# plot sentiment distribution for positive and negative reviews

import seaborn as sns

for x in [0, 1]:
    subset = reviews_df[reviews_df['is_bad_review'] == x]
    
    # Draw the density plot
    if x == 0:
        label = "Good reviews"
    else:
        label = "Bad reviews"
        
sns.distplot(subset['compound'], hist = False, label = label)
```
